```
# Q5. What percent times KKR decides to bat after winning the toss
```

```
s=df[(df['team1']=='Kolkata Knight Riders') | (df['team2']=='Kolkata Knight Riders')].id.count()
```

```
t=df[(df['toss_decision']=='bat') & (df['toss_winner']=='Kolkata Knight Riders')].id.count()
```

```
(t/s)*100
```
