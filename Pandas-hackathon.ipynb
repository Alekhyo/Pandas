**Problem 3**: Write a python function that will take batch name and a date range as input and outputs the most consistent student in that given daterange and batch

**Note**: Come up with your definition of consistency

**Marks - 20** 

```
import datetime
now = datetime.datetime.now()

b=list(batch_users['batch_name'].unique())
sd=list(batch_users['evaluation_date'].unique())

def validate(date_text):
    try:
        datetime.datetime.strptime(date_text, '%Y-%m-%d')
        return True
    except:
        return False

def consistent(batch_name,start_date,end_date):      #consistency=40% of score + 20% of winner_score + 20% of gyan + 20% of myday - 2*late_submission    
    d=batch_users[batch_users['batch_name']==batch_name]
    
    score=(d[(d['evaluation_date']>=start_date) & (d['evaluation_date']<=end_date)]).drop_duplicates('e_id')
    score=score.groupby('user_id')['score'].sum()*0.40
    
    winner=(d[(d['evaluation_date']>=start_date) & (d['evaluation_date']<=end_date)]).drop_duplicates('e_id')
    winner=winner.groupby('user_id')['winner'].count()*0.20
    
    gyan=(d[(d['evaluation_date']>=start_date) & (d['evaluation_date']<=end_date)]).drop_duplicates('gyan_id')
    gyan=gyan.groupby('user_id')['gyan_id'].count()*0.20

    myday=(d[(d['evaluation_date']>=start_date) & (d['evaluation_date']<=end_date)]).drop_duplicates('myday_id')
    myday=myday.groupby('user_id')['myday_id'].count()*0.20
    
    late=(d[(d['evaluation_date']>=start_date) & (d['evaluation_date']<=end_date)]).drop_duplicates('e_id')
    late=late.groupby('user_id')['late'].count()*2
    
    cons=(score+winner+gyan+myday-late).reset_index()
    cons.rename(columns={0:"Consistent"},inplace=True)
    return(cons.merge(users,left_on='user_id',right_on='user_id')[['user_id','fname','lname','Consistent']].sort_values('Consistent',ascending=False).head(1))
            
    
batch_name=input("Enter the batch name: ")
start_date=input("Enter the start date: ")
end_date=input("Enter the end date: ")

if (batch_name in b):
        if (len(start_date)==0 and len(end_date)==0):
            start_date=list(batch[batch['batch_name']==batch_name]['batch_start_date'])[0]
            end_date=(now.strftime("%Y-%m-%d"))
            print(consistent(batch_name,start_date,end_date))
            
        
        elif (validate(start_date)==True and validate(end_date)==True):
            if ((start_date>=sd[0]) and (end_date<=sd[-1])):
                print(consistent(batch_name,start_date,end_date))
            else:
                print("DATE OUT OF COURSE DURATION")
        
        else:
            print("INCORRECT DATE FORMAT")
        
else:
    print("INCORRECT BATCH NAME")
```

```
# Write your code here
b=list(batch_users['batch_name'].unique())
sd=list(batch_users['evaluation_date'].unique())

d=batch_users
    
score=(d[(d['evaluation_date']>=sd[0]) & (d['evaluation_date']<=sd[-1])]).drop_duplicates('e_id')
tasks=score.groupby('user_id')['e_id'].count()

score=(score.groupby('user_id')['score'].sum())/tasks

    
winner=(d[(d['evaluation_date']>=sd[0]) & (d['evaluation_date']<=sd[-1])]).drop_duplicates('e_id')
winner=(winner.groupby('user_id')['winner'].count()*10)/tasks
    
gyan=(d[(d['evaluation_date']>=sd[0]) & (d['evaluation_date']<=sd[-1])]).drop_duplicates('gyan_id')
gyan=gyan.groupby('user_id')['gyan_id'].count()

myday=(d[(d['evaluation_date']>=sd[0]) & (d['evaluation_date']<=sd[-1])]).drop_duplicates('myday_id')
myday=myday.groupby('user_id')['myday_id'].count()
    
late=(d[(d['evaluation_date']>=sd[0]) & (d['evaluation_date']<=sd[-1])]).drop_duplicates('e_id')
late=(late.groupby('user_id')['late'].count())*2
    
cons=(score+winner+gyan+myday-late+(tasks*10)).reset_index()                    #merit_points=(score/task)+(winnner*10)/task+gyan+myday-2*late+no.of task submission
cons.rename(columns={0:"Merit"},inplace=True)
(cons.merge(users,left_on='user_id',right_on='user_id')[['user_id','fname','lname','Merit']].sort_values('Merit',ascending=False).head(1))
```

```
def performance(batch_name):                                                    #performance=(total batch score/total_task) + (winner-late_submission)/total_task + profile_updates + (gyan+myday) + (no_of_tasks with more than 2 winners) - (no_of_task with more than 1 late submission)
  try: 
    best=batch_users[batch_users['batch_name']==batch_name]
    
    batch_score=best.drop_duplicates('e_id')['score'].sum()

    total_task=best.drop_duplicates('e_id')['e_id'].count()

    late=(best['late'].sum())

    winner=(best['winner'].sum())

    max_wins=best[best['winner']==1]
    max_wins=max_wins.groupby('task_number')['winner'].count().reset_index()
    max_wins=max_wins[max_wins['winner']>2].shape[0]
    
    
    max_late=best[best['late']==1]
    max_late=max_late.groupby('task_number')['late'].count().reset_index()
    max_late=max_late[max_late['late']>1].shape[0]

    profiles=(best.drop_duplicates('dp')['dp'].count()+best.drop_duplicates('github')['github'].count()+best.drop_duplicates('portfolio')['portfolio'].count()+best.drop_duplicates('medium')['medium'].count()+best.drop_duplicates('linkedin')['linkedin'].count()+best.drop_duplicates('youtube')['youtube'].count())*10

    others=(best.drop_duplicates('gyan_id')['gyan_id'].count()+best.drop_duplicates('myday_id')['myday_id'].count())

    total_score=(batch_score/total_task)+((winner-late)/total_task)+((profiles+others)/100)+(max_wins-max_late)

    return(total_score)
    
  except:
    print("Improper Batch Name")

batch_name=input("Enter batch name 1: ")
print(performance(batch_name))

batch_name=input("Enter batch name 2: ")
print(performance(batch_name))
```
